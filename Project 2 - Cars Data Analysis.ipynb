## 1) Instruction (For Data Cleaning)

• Find all Null Value in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
# fimd num of Null Value  

## sum() đếm số lượng giá trị True
## count() đếm tổng số phần tử trong một cột, bao gồm cả các giá trị không null.
## => Dùng sum()
data.isnull().sum()
 
```

```python
data.head(5)
```

```python
# fill null value by mean columns
data['EngineSize'] = data['EngineSize'].fillna(data['EngineSize'].mean())
data['Cylinders'] = data['Cylinders'].fillna(data['Cylinders'].mean())
data['Horsepower'] = data['Horsepower'].fillna(data['Horsepower'].mean() )
data['MPG_City'] = data['MPG_City'].fillna(data['MPG_City'].mean() )
data['MPG_Highway'] = data['MPG_Highway'].fillna(data['MPG_Highway'].mean() )
data['Weight'] = data['Weight'].fillna(data['Weight'].mean() )
# data['Length_1'] = data['Length_1'].fillna(data['Length_1'].mean() ) # tất cả data điều NAN
       
```
